***Q-9.Find out how many female passengers had survived and her age < 30 , show on pie chart and bar graph***

```python
mask2 = female_data['Age'] < 30
FSA = female_data[mask2]
SA = FSA['Survived'].value_counts()
SA2 = FSA.groupby(['Survived'], as_index = False)['Age'].count()
```

```python
plt.pie(x = SA, labels = ['Survived', 'Dead'], autopct= '%1.2f%%', shadow = True, startangle= 90, explode= (0,0.1), 
        wedgeprops={'edgecolor': 'black', 'linewidth': 2}, textprops= {'fontweight': 'bold', 'fontsize': 14},
        pctdistance= 0.7, colors = ['lightblue', 'lightgreen'])
plt.title('Female Younger than 30 yrs Survival (in %age)', fontweight = 'bold', fontsize = 18);
```

```python
from matplotlib.font_manager import FontProperties
```

```python
plt.figure(figsize=(5,7))
br1 = sns.barplot(x = 'Survived', y = 'Age', data = SA2, hue = 'Survived', zorder = 2)

for i in br1.containers:
    br1.bar_label(i, label_type = 'center', fontweight = 'bold', fontsize = 12)

plt.xlabel('Survival', fontsize = 12, fontweight = 'bold')
plt.ylabel('Count', fontweight = 'bold', fontsize = 12)
plt.xticks(ticks = [0,1], labels = ['Dead', 'Survived'], fontweight = 'bold')
plt.yticks(fontweight = 'bold')
handles, labels = br1.get_legend_handles_labels()
plt.legend(handles = handles, labels = ['Dead', 'Survided'], title ='Survival',
           prop = {'weight': 'bold'}, title_fontproperties = FontProperties(weight='bold'))
plt.grid(color = 'white', axis = 'y', zorder = 1, linewidth = 4)
plt.gca().set_facecolor(color = 'lightgrey')
plt.title('Female Younger than 30 yrs Survival', fontweight = 'bold', fontsize = 16, pad = 15)
br1.spines['left'].set_visible(False)
br1.spines['right'].set_visible(False)
br1.spines['top'].set_visible(False)
br1.tick_params(axis='y', length=0)
```

***Q-10.Find out how many Male passengers had survived and his age > 40, show on pie chart and bar graph.***

```python
mask3 = df[(df['Sex'] == 'male') & (df['Age'] > 40)]
msp = mask3.groupby(['Survived'], as_index = False)['Survived'].count()
```

```python
plt.figure(figsize = (6,6))
plt.pie(x = msp['Survived'], labels = ['Dead', 'Survived'], autopct= '%1.1f%%',shadow = True, explode = (0,0.1),startangle= 90, 
        wedgeprops= {'edgecolor': 'black', 'linewidth': 2}, textprops= {"fontweight": 'bold', 'fontsize': 12}, colors = ['orange','lightgreen'])
plt.title('Survival of Male Passsengers above age 40 yrs', fontsize = 18, fontweight = 'bold');
```

```python
plt.figure(figsize=(5,7))
ms = sns.barplot(x = msp.index, y = 'Survived', data = msp, hue = msp.index, zorder = 2)

for i in ms.containers:
    ms.bar_label(i, fontweight = 'bold')

handles = ms.get_legend_handles_labels()[0]
plt.legend(title = 'Survival', labels = ['Dead', 'Survived'], handles = handles, prop = {'weight': 'bold'}, title_fontproperties=FontProperties(weight = 'bold'))
plt.xlabel('Survival', fontweight = 'bold', fontsize = 12)
plt.ylabel('Count of Passengers', fontweight = 'bold', fontsize = 12)
plt.xticks(ticks = [0,1], labels = ['Dead', 'Survived'], fontweight = 'bold')
plt.yticks(fontweight = 'bold')
plt.grid(axis = 'y', color = 'white', linewidth = 3, zorder = 1)
plt.gca().spines[['left', 'top', 'right']].set_visible(False)
plt.gca().set_facecolor('lightgrey')
plt.title('Survival of Male Passengers above age 40 yrs', fontweight = 'bold', fontsize = 15, pad = 15);
```

***Q-11.Show age with 20 bins***

```python
plt.figure(figsize = (12,6))
age_data = df["Age"].dropna()

counts, bins = np.histogram(age_data, bins=20)   # Compute histogram bin counts manually
bin_centers = (bins[:-1] + bins[1:]) / 2

ax = sns.histplot(age_data, bins=20, kde=False, color = 'yellow', edgecolor = 'red', label = 'Hist', zorder = 2, alpha = 1)

plt.plot(bin_centers, counts, marker='o', linestyle='-', color='r', label='Age Distribution')

plt.title("Age Distribution of Titanic Passengers", fontsize = 16, fontweight = 'bold')
plt.xlabel("Age", fontweight='bold', fontsize=12)
plt.ylabel("Frequency in Count", fontweight='bold', fontsize=12)
plt.xticks(np.arange(0,84,4))
plt.grid(axis = 'y', color = 'white', linewidth = 3, zorder = 1)
plt.gca().spines[['left', 'top', 'right']].set_visible(False)
plt.gca().set_facecolor('lightgrey')
plt.tick_params(axis='both', length=0)
plt.legend();
```

***Q-12.Show age frequency with survived and not survived (Histogram)***

```python
plt.figure(figsize=(10,6))

# Histogram with hue based on survival
ad = sns.histplot(data=df, x="Age", hue="Survived", multiple="dodge", palette="Set1", edgecolor='black',zorder = 2,alpha = 1)

for i in ad.containers:
    ad.bar_label(i, fmt='%d', fontsize=9, color='black', padding = 4)

# Survived group
age_survived = df[df["Survived"] == 1]["Age"]
counts_s, _ = np.histogram(age_survived, bins=bins)
bin_centers = (bins[:-1] + bins[1:]) / 2 +1
plt.plot(bin_centers, counts_s, color='red', marker='o', linestyle='-', label='Survived (Line)', zorder=3)

# Not Survived group
age_notsurvived = df[df["Survived"] == 0]["Age"]
counts_ns, _ = np.histogram(age_notsurvived, bins=bins)
bin_centers2 = (bins[:-1] + bins[1:]) / 2 -1
plt.plot(bin_centers2, counts_ns, color='blue', marker='s', linestyle='--', label='Not Survived (Line)', zorder=3)

# Customize plot
plt.title("Age Distribution by Survival Status", fontsize=16, fontweight='bold')
plt.xlabel("Age", fontsize=12, fontweight='bold')
plt.ylabel("Frequency", fontsize=12, fontweight='bold')
plt.xticks(np.arange(0,84,4))
plt.legend(title="Survived", labels=["No", "Yes"])
plt.grid(axis='y', linestyle='--', zorder = 1)
plt.tick_params(axis='both', length=0)
plt.tight_layout();
```
